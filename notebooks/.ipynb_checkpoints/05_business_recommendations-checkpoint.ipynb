# Customer Segmentation Analysis: Recommendations for iFood
_Prepared by: Kiran Gosavi_

## 1. Project Objective
To analyze customer data, identify key behavioral patterns, and create distinct customer segments to inform a data-driven marketing strategy.

## 2. Key Findings
* The customer base has a wide income range, with most customers falling into the mid-to-high income bracket.
* Spending habits are highly correlated with income, and customers show a strong preference for high-value products like wine and meat.
* Overall campaign response rates are low, but customers who respond tend to be high spenders.

## 3. Customer Segments
Based on our K-Means analysis, we have identified three distinct segments:

| Segment | Average Income | Average Total Spend | Key Characteristics |
| :--- | :--- | :--- | :--- |
| **Mid-Tier Family Shopper** (Cluster 0) | ~$58,745 | ~$685 | High purchase frequency, with a high number of children. |
| **Low-Spender, Low-Engager** (Cluster 1) | ~$34,292 | ~$83 | Low income, low spending, and low engagement. |
| **High-Value, High-Spender** (Cluster 2) | ~$76,366 | ~$1,351 | Highest income, highest spending, and few to no children. |

## 4. Strategic Recommendations

### **A. For the High-Value, High-Spender Segment**
**Objective**: Retain and maximize the lifetime value of this segment.
* **Product Focus**: Offer premium product bundles (e.g., exclusive wine selections, gourmet meat packages).
* **Communication**: Use personalized, high-touch communication. Consider direct mail or private email invitations to exclusive events.
* **Campaigns**: Offer special campaigns for a new product line with a discount.

### **B. For the Mid-Tier Family Shopper Segment**
**Objective**: Increase basket size and promote family-oriented products.
* **Product Focus**: Promote meal kits, family-sized packs, and products with a lower price per unit.
* **Communication**: Use targeted email campaigns with family-friendly recipes or deals.
* **Campaigns**: Run "buy one, get one free" offers on products relevant to children.

### **C. For the Low-Spender, Low-Engager Segment**
**Objective**: Increase engagement and drive initial purchases.
* **Product Focus**: Focus on promoting special deals, discounts, and high-value-for-money products.
* **Communication**: Use low-cost channels like SMS campaigns.
* **Campaigns**: Offer a one-time, high-percentage discount on their first purchase to incentivize spending.


### Predictive Modelling

**Objective:** Build a predictive model to identify customers most likely to respond to a new marketing campaign.

***

### Key Features

* **`MntTotal` (Total Amount Spent):** Our most significant predictor.
* **`Age`:** Strong indicator of campaign response.
* **`Kids`:** Highly influential on purchasing behavior.

***

### Model & Performance

* **Model:** Logistic Regression
* **Key Performance Indicators (KPIs):**
    * **Accuracy:** 87%
    * **Precision:** 85%
    * **Recall:** 78%
    * **F1-Score:** 81%

***

### Key Observation & Recommendation

* **Observation:** The model achieved a **Recall of 78%**, meaning it is highly effective at identifying customers who will actually respond to a campaign. This is crucial for not missing out on high-value leads.
* **Recommendation:** Use the model to create a **"Hot List"** of the top-predicted responders. Focus your marketing budget on this targeted group to reduce costs and **increase campaign ROI by an estimated 20-30%** by avoiding mass-mailing to un-responsive customers.
